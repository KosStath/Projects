```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import pandas as pd

driver = webdriver.Chrome()


url = "https://www.cev.eu/national-team/european-league/2023/european-golden-league/men/statistics/#leader-lists"
driver.get(url)

time.sleep(3)

select_element = driver.find_element(By.XPATH, "/html/body/main/div/div/div[2]/div/div[1]/div/div/div/fieldset/div[2]/div[1]/select")

options = Select(select_element).options

dfs = {}

for option in options:
    
    Select(select_element).select_by_value(option.get_attribute("value"))

    
    time.sleep(3)

    
    table_xpath = "/html/body/main/div/div/div[2]/div/div[1]/div/div/div/fieldset/div[4]/table"
    table = driver.find_element(By.XPATH, table_xpath)

    
    headers = [th.text.replace(' ', '') for th in table.find_elements(By.TAG_NAME, "th")]
    
    
    rows = table.find_elements(By.TAG_NAME, "tr")
    
    
    data = []

    
    for row in rows[1:]:  
        columns = row.find_elements(By.TAG_NAME, "td")
        player_data = [col.text for col in columns]
        data.append(player_data)

    
    df = pd.DataFrame(data, columns=headers)
    
    
    df.set_index(headers[0], inplace=True, drop=True)

    
    dfs[option.text] = df
    
    
    df.to_csv(f"{option.text}_Volley_GL_Men_'23.csv")


driver.quit()

# Access each DataFrame using the select option as the key
"""for option, df in dfs.items():
    print(f"Data for {option}:")
    print(df)
    print("\n")"""

```
